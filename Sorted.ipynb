```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("C:/Users/Admin/Documents/MamathaKPYTHON/loanclean.csv")
df['loan_status'].value_counts()
dfchargedoff = df[df['loan_status'] == 'Charged Off']
```

```python
###Univariate Analysis, Users who have chargedoff  
###loan amt less than 10K are more prone to charged off.
dfchargedoff.loan_amnt.plot.hist()

```

```python
## see if purpose has more defaults 
sns.countplot(x = 'purpose', data = df, palette = "Blues");
plt.xticks(rotation=90)
plt.show()
```

```python
####Loan_Amount VS Recovery
#####AS and when loan amount is increasing recovery is seen less for charged off account.. 
sns.pairplot(dfchargedoff[['recoveries','loan_amnt','annual_inc']])
plt.show()
```

```python

##people is rented are more prone to chargedoff
sns.countplot(x = 'home_ownership', data = dfchargedoff, palette = "Blues");
plt.show()
```
