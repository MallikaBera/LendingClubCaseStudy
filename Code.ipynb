Observation 2 : A lot of columns are empty, while a lot may have more than 50% values missing
Action: Identify all columns which has more than 50% missing values and drop them from the dataset.

```python
# Check for columns with more than 50% missing values(NA)

loan = loan.loc[:, (loan.isna().sum() < (loan.shape[0]/2))]
loan

```

```python
# Check for columns with more than 50% missing values(0)
loan = loan.loc[:, ((loan == 0).sum() <(loan.shape[0]/2))]
loan
```
